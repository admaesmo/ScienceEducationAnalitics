### Unificamos los datos

````python
#primero creamos las columas que almacenaran los datos entrantes de encuesta.csv
columnas = ['Licenciatura', 'Profesional', 'Especializacion CN', 'Especilizacion Ed',
           'Maestria CN', 'Maestria Ed', 'Doctorado CN', 'Doctorado Ed']

for columna in columnas:
  EncuestaMod1[columna] = 0
````

````python
# Elimina espacios en blanco en la columna 'titulo' de Encuesta
Encuesta['Titulo'] = Encuesta['Titulo'].str.strip()

# Elimina espacios en blanco en la columna 'titulo' de EncuestaMod1
EncuestaMod1['Titulo'] = EncuestaMod1['Titulo'].str.strip()
````

````python
# ahora iteremos sobre el dataframe EncuestaMod1 para usar titulo como llave foranea y recibir los datos
titulos = ['Licenciatura', 'Profesional', 'Especializacion CN', 'Especilizacion Ed','Maestria CN', 'Maestria Ed', 'Doctorado CN', 'Doctorado Ed']

# Crea un diccionario para mapear los títulos
mapeo_titulos = {Titulo: Titulo for Titulo in titulos}

# Itera sobre las filas de EncuestaMod1
for index, row in EncuestaMod1.iterrows():
  # Busca el título correspondiente en Encuesta
  titulo_encuesta = Encuesta.loc[Encuesta['Titulo'] == row['Titulo'], titulos].values

  # Si se encuentra el título, actualiza las columnas en EncuestaMod1
  if len(titulo_encuesta) > 0:
    for Titulo in titulos:
      EncuestaMod1.loc[index, mapeo_titulos[Titulo]] = titulo_encuesta[0][titulos.index(Titulo)]
````
