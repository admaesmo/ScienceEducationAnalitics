usar el método str.contains() de pandas para crear nuevas columnas basadas en la presencia de un valor específico en la columna "grado".

Primero, crea las nuevas columnas y asígnales el valor 0:

```python
grados = ['6', '7', '8', '9', '10', '11']
for grado in grados:
  dfCompleto[f'grado {grado}'] = 0
```

itera sobre las filas del DataFrame y marca las nuevas columnas con 1 si la columna "grado" contiene el grado correspondiente:

```python
for index, row in dfCompleto.iterrows():
  for grado in grados:
    if f'{grado}º' in row['grado']:
      dfCompleto.loc[index, f'grado {grado}'] = 1
```

```python
# Define las columnas de interés
titulos = ['Licenciatura', 'Profesional', 'Especializacion CN', 'Especilizacion  Ed','Maestria CN', 'Maestria Ed', 'Doctorado CN', 'Doctorado Ed']

# Crea nuevas columnas para cada título y para "Varios títulos"
for titulo in titulos:
  Encuesta[f'Tiene {titulo}'] = 0
Encuesta['Tiene Varios Títulos'] = 0

# Recorre cada fila del DataFrame
for index, row in Encuesta.iterrows():
  titulos_docente = [titulo for titulo in titulos if row[titulo] == 1]
  if len(titulos_docente) == 1:
    Encuesta.loc[index, f'Tiene {titulos_docente[0]}'] = 1
  elif len(titulos_docente) > 1:
    Encuesta.loc[index, 'Tiene Varios Títulos'] = 1

# Ahora puedes usar las nuevas columnas para crear histogramas o gráficos de calor
# Por ejemplo, para crear un histograma de la columna 'Tiene Licenciatura':
Encuesta['Tiene Licenciatura'].hist()

# O para crear un gráfico de calor de la correlación entre las nuevas columnas:
import seaborn as sns
sns.heatmap(Encuesta[['Tiene Licenciatura', 'Tiene Profesional', 'Tiene Varios Títulos']].corr()) # Changed df to Encuesta
```

primero se define una lista con las columnas que se usarán para el gráfico. Luego, calcula el número de docentes para cada columna sumando los valores (que son 1 o 0) y la proporción dividiendo el número de docentes por el total de docentes en el DataFrame.

Finalmente, crea un gráfico de barras con dos ejes Y: uno para el número de docentes (en azul) y otro para la proporción de docentes (en rojo). El gráfico muestra la información de forma clara y concisa, permitiendo comparar el número y la proporción de docentes para cada título académico.\

```python
# @title Default title text
# ... (código para crear las nuevas columnas)

# Define las columnas para el gráfico
columnas_grafico = ['Tiene Licenciatura', 'Tiene Profesional', 'Tiene Especializacion CN',
                   'Tiene Especilizacion  Ed', 'Tiene Maestria CN', 'Tiene Maestria Ed',
                   'Tiene Doctorado CN', 'Tiene Doctorado Ed']

# Calcula el número de docentes para cada columna
conteo_docentes = Encuesta[columnas_grafico].sum()

# Calcula la proporción de docentes para cada columna
proporcion_docentes = conteo_docentes / len(Encuesta)

# Crea el gráfico de barras
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.bar(conteo_docentes.index, conteo_docentes.values, color='tab:blue')
ax1.set_xlabel('Título Académico')
ax1.set_ylabel('Número de Docentes', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')
plt.xticks(rotation=45, ha='right')

ax2 = ax1.twinx()
ax2.plot(proporcion_docentes.index, proporcion_docentes.values, color='tab:red', marker='o')
ax2.set_ylabel('Proporción de Docentes', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

plt.title('Número y Proporción de Docentes por Título Académico')
plt.tight_layout()
plt.show()
```
